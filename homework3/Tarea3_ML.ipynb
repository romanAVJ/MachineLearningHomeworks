# Data set: circles


Usaremos el conjunto de datos de _sklearn_ __make circles__ para los ejercicios de la actual tarea

```python
# all libraries to import
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
import numpy as np

## personal functions
# deep neural network
import dnn
# kernels

#utils
from utils import plot_decision_boundary
```

```python
# import data
X, y = make_circles(2000, noise = .1, random_state = 123, factor = .4)
m = X.shape[1]
n = X.shape[0]

# look data
plt.plot(X[y==0,0], X[y==0,1], "bo")
plt.plot(X[y==1,0], X[y==1,1], "ro")
plt.title('Número de observaciones: ' + str(n))
```

```python
# look data
print('Los datos X están dados por: \n', X[:5])
print('\n\nLas etiquetas Y están dados por: \n', y[:5])
```

# Red Neuronal 

Se implementará una red neuronal de tres capas de pesos: 1 capa de entrada, 2 ocultas y una de salida.
La forma de la matriz de características es de la forma
$$X \in \mathbb{R}^{m \times n},$$

donde $m$ es el número de características y $n$ el número de observaciones. En este caso, no se incluye la característica $x_0^{[l]} = 1$ para el intercepto, si no que en cada capa $l$ se calcula de forma directa. Esta notación es sugerida por Andrew NG, fundador de Google DeepLearning.AI, en su curso impartido en Coursera.

De igual forma $Y \in \mathbb{R}^{1 \times n}$, es un vector renglón.

```python
# reshape data
X_dnn = X.T
y_dnn = y.reshape((1, n))

# look data
print('Los datos X transpuestos están dados por: \n', X_dnn)
print('\n\nLas etiquetas Y reformadas están dadas por: \n', y_dnn)
```

```python
y_dnn
```

## Arquitectura de la red neuronal

Se usará

- X_dnn
- y_dnn
- layer_dims = (m, 3, 3, 1). Se escogieron jugando en Tensorflow [playground](https://playground.tensorflow.org/#activation=relu&batchSize=10&dataset=circle&regDataset=reg-plane&learningRate=0.1&regularizationRate=0&noise=0&networkShape=3,3&seed=0.55412&showTestData=false&discretize=false&percTrainData=50&x=true&y=true&xTimesY=false&xSquared=false&ySquared=false&cosX=false&sinX=false&cosY=false&sinY=false&collectStats=false&problem=classification&initZero=false&hideText=false)
- learning rate = 0.1
- epochs = 1000



A continuación ajustamos la red neuronal con las capas especificadas anteriormente.

```python
# fit dnn
# values 
layers = [m, 3, 3, 1]
lr = 1

# dnn
params, costs = dnn.dnn(X_dnn, y_dnn,
                      layers_dims=layers,
                      learning_rate=lr,
                      epochs=3000)
```

### Gráfica de los costos por época

Observamos el costo conforme pasan las épocas para la red neuronal. Vemos que está entrenando muy bien.

```python
# look cost
epochs = list(costs.keys())
cost = list(costs.values())

# look 
plt.plot(epochs, cost)
plt.ylabel('costo')
plt.xlabel('epochs (por cientos)')
plt.xticks(rotation=90)
plt.title("Tasa de aprendizaje =" + str(lr))
plt.show()
```

## Ajuste de la red neuronal

### Error de ajuste

Vemos el error de ajuste para los datos ajustados

```python
# predict
yhat_dnn = dnn.predict(X_dnn, params)

# error
err_dnn = 1 - np.mean(yhat_dnn == y_dnn)

print("La red neuronal tiene un error de ajuste de {0}% ".format(err_dnn))
```
